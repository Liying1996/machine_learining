### 2. iris的基本信息

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris

data = load_iris()

# type(data)
# 分类情况
print('分类情况为：')
print(data.target) 

# 转换数据格式
iris = pd.DataFrame(data=data.data, columns=data.feature_names)
print('数据集前五行为：')
print(iris.head(5)) 

# 基本信息
print('基本信息为：')
iris.info()
print('数据统计信息为：')
iris.describe()
```

---
### 3. 可视化iris
下面将用seaborn可视化。

```python
# 构建新的数据集iris_all，加上分类的信息
iris_all = iris.copy()
target = [data.target_names[i] for i in data.target]
iris_all['type'] = target
print(iris_all.head(5))

# pairplot
# help(sns.pairplot)
sns.pairplot(data=iris_all, hue='type') # density
sns.pairplot(data = iris_all, hue = 'type', diag_kind = 'hist')
```

```python
# 不同的feature之间基本都可以将三种花分开，进一步用boxplot看一下
for name in iris_all.columns[:-1]:
    sns.boxplot(x='type', y=name, saturation=0.5,palette='pastel', data=iris_all)
    plt.title(name)
    plt.show()
```
